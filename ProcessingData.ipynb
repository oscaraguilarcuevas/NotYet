# Creating functions for parsing dates

```python
date
```

```python
def extract_year(date_string):
    year_pattern = r"\d{4}$"
    year, = re.findall(year_pattern, date_string)
    return year
```

```python
extract_year(date)
```

```python
def extract_day(date_string):
    day_pattern = r"^.{8}(\d+)\s"
    day, = re.findall(day_pattern, date_string)
    return day
```

```python
extract_day(date)
```

```python
def extract_month(date_string):
    month_pattern = r"^\w{3}\s(\w{3})\s"
    month_string, = re.findall(month_pattern, date_string)
    
    #translate string to MM representation
    months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
    month_number = months.index(month_string) + 1
    month_no = ("0" if month_number < 10 else "") + str(month_number)
    
    return month_no
```

```python
extract_month(date)
```

```python
def extract_date(date_string):
    return "-".join([extract_year(date_string), extract_month(date_string), extract_day(date_string)])
```

```python
extract_date(date)
```

```python
conversions = {"city": lambda x: x.capitalize(),  
               "zip": int,           
               "beds": int,
               "price": int,
               "baths": int,
               "sq__ft": int,
               "latitude": float,
               "longitude": float,
               "sale_date": extract_date,}

def process_row(row, conversion_mapping):
    new_row = {}
    
    for key, value in row.items():
        if key in conversion_mapping:
            new_row[key] = conversion_mapping[key](value)
        else:
            new_row[key] = value
            
    return new_row  # Pay attention to the indentation
```

```python
process_row(sample, conversions)
```

```python
[process_row(sample, conversions) for row in data]
```

# Writing processed dataset to a csv file

```python
clean = [process_row(sample, conversions) for row in data]
```

```python
clean[0].keys()
```

```python
with open("sacramento_clean.csv", "w") as outfile:
    fieldnames = list(clean[0].keys())
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for row in clean:
        writer.writerow(row)
```
